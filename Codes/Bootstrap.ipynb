```python
import pandas as pd
import numpy as np
from sklearn.utils import resample

input_filename = 'input_data.csv'
output_filename = 'output_data.csv'
n_iterations = 1000

data = pd.read_csv(input_filename)
numeric_cols = data.select_dtypes(include=[np.number]).columns
results = {}
for col in numeric_cols:
    median = data[col].dropna().median()
    mean = data[col].dropna().mean()
    std = data[col].dropna().std()
    q1 = data[col].dropna().quantile(0.25)
    q3 = data[col].dropna().quantile(0.75)
    iqr = q3 - q1
    n=data[col].dropna().count()
    bootstrapped_medians = []
    for i in range(n_iterations):
        sample = resample(data[col].dropna(), replace=True, n_samples=len(data[col].dropna()))
        bootstrapped_median = np.median(sample)
        bootstrapped_medians.append(bootstrapped_median)
    ci_lower = np.percentile(bootstrapped_medians, 2.5)
    ci_upper = np.percentile(bootstrapped_medians, 97.5)
    results[col] = {'Median': median,
                    'Mean': mean,
                    'Std': std,
                    'Q1': q1,
                    'Q3': q3,
                    'IQR': iqr,
                    'No. of Obs': n,
                    'CI Lower': ci_lower,
                    'CI Upper': ci_upper}
results_df = pd.DataFrame.from_dict(results, orient='index')
results_df.to_csv(output_filename)
```
